```python
dayOne_data = open('/AOC/AOC_Day_1.txt', 'r')
```

```python
#part - 1
all_numbers = []
for line in dayOne_data:
  all_numbers.append("")
  for character in line:
    if character.isdigit():
      all_numbers[-1] = all_numbers[-1] + str(character)
print(all_numbers)
```

```python
#part - 1
updated_all_numbers = []
for number in all_numbers:
  print(len(number))
  if 2 < len(number) or len(number) == 1:
    updated_all_numbers.append(str(number[0]) + str(number[-1]))
  else:
    updated_all_numbers.append(number)
print(updated_all_numbers)
```

```python
#part - 1
final_sum = sum([int(each_line_calibration) for each_line_calibration in updated_all_numbers])
print(final_sum)
```
